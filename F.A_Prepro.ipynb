```python
import pandas as pd 
import numpy as np 
```

# Balance Sheet Annualy

```python
balance_sheet_df=pd.read_excel(r'Financial Statement(PLTR).xlsx',sheet_name='Balanace Sheet Annualy')
balance_sheet_df=balance_sheet_df.iloc[2:,1:-1].reset_index(drop=True)
balance_sheet_df
```

```python
bs_df1=balance_sheet_df.transpose()
bs_df1 = bs_df1.reset_index(drop=False)
bs_df1.columns= bs_df1.iloc[0]
bs_df1=bs_df1.drop(0)
bs_df1
```

# Cash Flow Statement

```python
cf_statement=pd.read_excel('Financial Statement(PLTR).xlsx',sheet_name='Cash Flow Annualy')
cf_statement=cf_statement.iloc[2:,1:-1].reset_index(drop=True)
cf_statement
```

```python
cf_df=cf_statement.transpose()
cf_df=cf_df.reset_index(drop=False)
cf_df.columns=cf_df.iloc[0]
cf_df=cf_df.drop(0)
cf_df
```

```python
bs_anal=bs_df1[['Fiscal Year']].copy()
bs_anal
```

```python
# Calculate the Working Capital
current_asset=bs_df1['Total Current Assets'].str.replace(',','').astype(float)
current_Liabilities=bs_df1['Total Current Liabilities'].str.replace(',','').astype(float)
working_capital=current_asset -current_Liabilities
bs_anal['Working capital']=working_capital
bs_anal
```

```python
#Current Ratio
current_ratio=current_asset / current_Liabilities
bs_anal['Current ratio']= current_ratio
bs_anal
```

```python
# Debt to Equity Ratio:
total_debt=bs_df1['Total Debt'].str.replace(',', '').astype(float)
shareholders_equity=bs_df1['Shareholders\' Equity'].str.replace(',','').astype(float)
debt_to_equity=total_debt / shareholders_equity
bs_anal['Debt to Equity Ratio']=debt_to_equity
bs_anal
```

```python
receivable=bs_df1['Receivables'].str.replace(',','').astype(float)
payable=bs_df1['Accounts Payable'].str.replace(',','').astype(float)
bs_anal['Receivable']= receivable
bs_anal['Payable']= payable
```

# Income Statement Annualy

```python
is_annu=pd.read_excel(r'Financial Statement(PLTR).xlsx',sheet_name='Income Statement Annualy')
is_annu=is_annu.iloc[2:,1:-1].reset_index(drop=True)
is_annu
```

```python
# Transpose Data Frame
is_df=is_annu.transpose()
is_df= is_df.reset_index(drop=False)
is_df.columns=is_df.iloc[0]
is_df=is_df.drop(0)
is_df
```

```python
is_anal=is_df[['Fiscal Year']].copy()
is_anal
```

```python
revenue=is_df['Revenue'].str.replace(',','').astype(float)
cogs=is_df['Cost of Revenue'].str.replace(',','').astype(float)
dso=(bs_anal['Receivable']/ revenue)*365
dpo=(bs_anal['Payable']/ cogs)*365
bs_anal['DSO']=dso
bs_anal['DPO']=dpo
bs_anal['CCC']=dso-dpo
bs_anal
```

```python
tota_assets=bs_df1['Total Assets'].str.replace(',','').astype(float)
bs_anal['Total Assets']=tota_assets
bs_anal['Begenning Total Assets']=bs_anal['Total Assets'].shift(-1,fill_value=bs_anal['Total Assets'].iloc[-1])
Average_assets=(bs_anal['Begenning Total Assets']+bs_anal['Total Assets'])/2
assets_turnover=revenue/ Average_assets
bs_anal['Total Assets Turnover']=assets_turnover
bs_anal
```

```python
bs_anal['Total Equity']=shareholders_equity
bs_anal['Begenning Total Equity']=bs_anal['Total Equity'].shift(-1,fill_value=bs_anal['Total Equity'].iloc[-1])
average_equity=(bs_anal['Begenning Total Equity']+bs_anal['Total Equity'])/2
bs_anal['Equity Turnover']=revenue/average_equity
```

# Profitability Ratios

```python
prof_df=bs_df1[['Fiscal Year']].copy()
prof_df['Gross Profit']= revenue - cogs
prof_df['Revenue']=revenue
net_inc=is_df['Net Income'].str.replace(',','').astype(float)
ebitda=is_df['EBITDA'].str.replace(',','').astype(float)
ebit=is_df['EBIT'].str.replace(',','').astype(float)
ebt=is_df['Pretax Income'].str.replace(',','').astype(float)
prof_df['Net Income']=net_inc
prof_df['EBITDA']=ebitda
prof_df['EBIT']=ebit
prof_df['Gross Profit Margin %']=(prof_df['Gross Profit']/prof_df['Revenue'])*100
prof_df['Operating profit margin %']= (ebit / revenue)*100
prof_df['Pre-tax margin %']=(ebt / revenue)*100
prof_df['Net profit margin %']=(net_inc/revenue)*100
prof_df['Return on assets (ROA)%']=(net_inc /Average_assets) *100
prof_df['Return on equity (ROE) %']=(net_inc/average_equity)*100
profitability_ratios_df=prof_df.copy()
profitability_ratios_df
```
